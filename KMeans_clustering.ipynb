```python
from sklearn.cluster import KMeans
```

```python
inertia_list = []
for k in np.arange(1,15):
    kmn = KMeans(n_clusters = k)
    kmn.fit(df[['Faculty of Graduation','Year of Graduation','Year of First Gift','Year of Last Gift','Largest Gift','# of Clicks in the Past Month','Lifetime Giving']])
    inertia_list.append(kmn.inertia_)
    
inertia_list
```

```python
plt.plot(np.arange(1,15), inertia_list, 'ro-')
plt.xlabel("number of clusters")
plt.ylabel("inertia")
plt.show()
```

```python
kmn= KMeans(n_clusters = 3)

kmn.fit(df[['Faculty of Graduation','Year of Graduation','Year of First Gift','Year of Last Gift','Largest Gift','# of Clicks in the Past Month','Lifetime Giving']])
labels = kmn.predict(df[['Faculty of Graduation','Year of Graduation','Year of First Gift','Year of Last Gift','Largest Gift','# of Clicks in the Past Month','Lifetime Giving']])
```
